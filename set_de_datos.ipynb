```python
# para quitar los archivos vasio se usa el siguiente codigo

datos_na = datos.dropna()
```

```python
datos_na.info()
```

```python
# Convertir la caracteristica categorica a numerica
#proximidad al oceano

datos_na["ocean_proximity"]
```

```python
datos_na["ocean_proximity"].value_counts()
```

```python
# Dummies / One-hot Encoding
# NEAR BAY  INLAND    NEAR OCEAN
# 1          0           1
# 0          0           1

dummies = pd.get_dummies(datos_na["ocean_proximity"], dtype=int)
```

```python
# aqui se agregan las columnas de 0 y 1


datos_na.join(dummies)
```

```python
datos_na.head()
```

```python
# para eliminar un renglon ejemplo ocean_proximity
datos.drop('ocean_proximity', axis=1)
```

```python
#confirmemos como quedaron nuestros datos
datos_na.head()
```

```python
# Unir las columnas dummy al DataFrame datos_na
datos_con_dummies = datos_na.join(dummies)

# Eliminar la columna original 'ocean_proximity' del nuevo DataFrame
datos_final = datos_con_dummies.drop('ocean_proximity', axis=1)

# Confirmar los datos
display(datos_final.head())
```

```python
# Analisis, nuevas caracteristicas
# correlaciones
datos_final.corr()
```

```python
## grafica
sb.set(rc={'figure.figsize':(15,8)})
sb.heatmap(datos_final.corr(), annot=True, cmap="coolwarm")
```
