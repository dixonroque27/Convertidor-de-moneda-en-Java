```python
import pandas as pd
datos = pd.read_csv('pesoaltura.csv')

```

```python
import seaborn as sb
datos.columns = datos.columns.str.strip()
sb.scatterplot(x ="PESO (Lbs)", y="ALTURA (pie)", data=datos)
```

```python
datos.columns = datos.columns.str.strip()
X = datos["PESO (Lbs)"]
Y = datos["ALTURA (pie)"]
```

```python
X.values
```

```python
Y.values
```

```python
X1 = X.values.reshape(-1,1)
Y1 = Y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X1,Y1)
```

```python
while True:
  calcular_peso = int(input("Introduzca su peso para conocer su estatura promedio : "))
  prediccion = modelo.predict([[calcular_peso]])
  print("Su estatura promedio es de : ", prediccion)
```
